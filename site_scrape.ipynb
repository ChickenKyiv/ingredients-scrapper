```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://newleafwellness.biz/2015/10/19/31-gluten-free-dairy-free-crockpot-freezer-meals/?removed_item=1#_a5y_p=4862964'
```

```python
response = requests.get(url)
```

```python
data = response.text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for link in soup.find_all(["a", "href"],rel="noopener", target='_blank', href=True):
    print(link['href'])
```

```python
url_list = list()
for link in soup.find_all(["a", "href"],rel="noopener", target='_blank', href=True):
    url_list.append(link['href'])
```

```python
data_ingre = {}
for link in url_list:
    recipie_ingre = []
    response2 = requests.get(link)
    data2 = response2.text
    soup2 = BeautifulSoup(data2, 'html.parser')
    title2 = soup2.find("title").text.split('–')[0]
    div_tag = soup2.find_all("div", class_="tasty-recipes-ingredients")
    for line in div_tag:
        for li_tag in line.find_all("li"):
            recipie_ingre.append(li_tag.text)
    data_ingre[title2] = recipie_ingre
```

```python
data_ingre
```

```python
data_dict = { k : v for k,v in data_ingre.items() if  v }
```

```python
data_dict
```

```python
imporved_data = dict()
for key, value in data_dict.items():
    new_data_list = []
    for ingredient in value:
        data_list = ingredient.split(' ')
        amount = ' '.join(data_list[:2])
        ingre = ' '.join(data_list[2:])
        new_value = (amount, ingre)
        new_data_list.append(new_value)
    imporved_data[key] =  new_data_list
```

```python
df = pd.DataFrame.from_dict(imporved_data, orient='index').transpose()
```
